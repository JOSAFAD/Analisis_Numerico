9. Escribir una función que reciba un número entero positivo y devuelva su factorial.

```
def factorial(numero):
    producto = 1
    i = 2
    while i <= numero:
        producto = producto * i
        i=i+1
    return producto

numero = int(input("Ingrese un numero: "))
res = factorial (numero)

print("El factorial es: ",res)
```
